### Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import the Datasets

```python
dataset = pd.read_csv("Position_Salaries.csv")
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Train & Visualize Different Models on Dataset

### Model Training - Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x, y)
```

### Visualization - Simple Linear Rgression

```python
plt.scatter(x, y, color='red')
plt.plot(x, regressor.predict(x), color='blue')

plt.title("Position vs Salaries - Simple Linear Regression")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

### Model Training - Polynomial Linear Regression with degree 2

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)

lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly, y)
```

### Visualization - Polynomial Linear Regression with degree 2

```python
plt.scatter(x, y, color='red')
plt.plot(x, lin_reg2.predict(x_poly), color='blue')

plt.title("Position vs Salaries - Polynomial Linear Regression (degree 2)")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

### Model Training - Polynomial Linear Regression with degree 4

```python
poly_reg3 = PolynomialFeatures(degree=4)
x_poly3 = poly_reg3.fit_transform(x)

lin_reg3 = LinearRegression()
lin_reg3.fit(x_poly3, y)
```

### Visualization - Polynomial Linear Regression with degree 4

```python
plt.scatter(x, y, color='red')
plt.plot(x, lin_reg3.predict(x_poly3), color='blue')

plt.title("Position vs Salaries - Polynomial Linear Regression (degree 4)")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
